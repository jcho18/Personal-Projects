## change to datetime

```python
from datetime import datetime
year = 2020
month = 'March'

s = f"{year} {month} 1"
d = datetime.strptime(s, '%Y %B %d')
d.strftime('%Y-%m-%d')
```

```python
def find_datetime(row):
    
    year = row['year']
    month = row['month']
    
    s = f"{year} {month} 1"
    d = datetime.strptime(s, '%Y %B %d')
    
    return d.strftime('%Y-%m-%d')
```

```python
row = {'year': 1949, 'month': 'January'}
find_datetime(row)
```

```python
import pandas as pd
df['date'] = df.apply(find_datetime, axis = 1)
df['date'] = pd.to_datetime(df['date'])
df['date'].head(1)
```

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
plt.figure(figsize = (15, 5))
sns.lineplot(data = df, x = 'date', y = 'passengers')
```

## Time series trend

```python
df_date = df[['date', 'passengers']].set_index('date')
df_date.head()
```

### Additive model

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomp = seasonal_decompose(df_date, model = 'additive')
d = decomp.plot()
```

### Muliplicative model

```python
decomp = seasonal_decompose(df_date, model = 'multiplicative')
d = decomp.plot()
```
