```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/jay/Desktop/Personal-Projects/dataset/covid_19_clean_complete.csv", parse_dates=['Date'])
```

```python
df.head()
```

```python
df = df.rename(columns={"Province/State": "province_state", "Country/Region": "country_region"})
```

```python
df['Active'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
```

```python
df.head()
```

```python
df = df[df['Date'] == max(df['Date'])].reset_index()
```

```python
df.groupby('country_region')['Confirmed', 'Deaths', 'Recovered'].sum().reset_index()
```

```python
df.to_csv('/Users/jay/Desktop/Personal-Projects/dataset/covid-19-cleaned.csv', index = False)
```
